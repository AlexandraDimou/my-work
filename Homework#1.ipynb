```python
#Using pandas to_datetime in order to convert column Date from dtype object to dtype datetime64[ns]
df['Date'] = pd.to_datetime(df['Date'], utc= True)
```

```python
df.info()
```

```python
#Adding a column named day_of_week where the day of the week is displayed
df['day_of_week'] = df['Date'].dt.day_name()
```
